```python
import urllib.request
from bs4 import BeautifulSoup as bs
import re 
import pandas as pd 

```

```python
# load html from a url 
page = urllib.request.urlopen("https://www.newegg.ca/p/pl?N=100007708%20601408872&PageSize=96")
soup = bs(page)

# Display formatted html code 
print(soup.prettify())
```

```python
gpus = soup.find_all('div', class_='item-container') # find all divs with class item-container which are the gpus on the webpage 
price_pattern = r'\$([\d,]+(?:\.\d{2})?)' # regex pattern extract the price 

for item in gpus:
    print("******************************************************************************************")
    print(item.find('a', class_='item-title').text)
    print(item.find('a', class_='item-title')['href'])
    print()
    price_text = item.find('li', class_='price-current').text
    price_match = re.search(price_pattern, price_text)
    if price_match:
        price = price_match.group(1)  # Extract the matched price without the dollar sign
        price = price.replace(',', '')  # Remove commas from the price string
        print(price)
    else:
        print("Price not found.")
    print("******************************************************************************************")
    print("\n")
```

```python

# Lists to store the scraped data in
titles = []
links = []
prices = []
```

```python

# Extract data from individual gpu container and appending to the lists 
for item in gpus:
    titles.append(item.find('a', class_='item-title').text)
    links.append(item.find('a', class_='item-title')['href'])
    price_text = item.find('li', class_='price-current').text
    price_match = re.search(price_pattern, price_text)
    if price_match:
        price = price_match.group(1)  # Extract the matched price without the dollar sign
        price = price.replace(',', '')  # Remove commas from the price string
        prices.append(float(price))
    else:
        prices.append(float('nan'))  # Append NaN for missing prices
```

```python
# Creating a dataframe from the lists as columns
df = pd.DataFrame({'Title': titles, 'Link': links, 'Price': prices})
```
